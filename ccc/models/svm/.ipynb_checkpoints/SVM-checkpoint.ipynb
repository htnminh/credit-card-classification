# Import necessary library

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

from sklearn.ensemble import RandomForestClassifier
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.utils.validation import column_or_1d
from sklearn.impute import SimpleImputer
from sklearn.svm import LinearSVC
from sklearn.svm import SVR
from sklearn.svm import SVC, NuSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split,RepeatedStratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score,confusion_matrix, f1_score, classification_report, fbeta_score, make_scorer,plot_roc_curve
from sklearn.metrics import ConfusionMatrixDisplay
from imblearn.over_sampling import SMOTE
from sklearn.feature_selection import RFE,RFECV
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingGridSearchCV
```

# Data collection

```python
def getDataset(file: str):
    for i in range(len(os.path.dirname(os.getcwd()))):
        if os.path.dirname(os.getcwd())[-i] == 'c':
            file = os.path.dirname(os.getcwd())[:-i+1] + "\\data\\"+file
            break
    return file
X_train_prep = pd.read_csv(getDataset('X_train_prep.csv')) 
X_test_prep = pd.read_csv(getDataset('X_test_prep.csv'))
y_train_prep = pd.read_csv(getDataset('y_train_prep.csv'))
y_test_prep = pd.read_csv(getDataset('y_test_prep.csv'))

X_test,X_val, y_test, y_val = train_test_split(X_test_prep, y_test_prep, test_size= 0.4, random_state=1)
```

# Data Visualize

```python
pca = PCA(n_components=2)
X_train1 = pd.DataFrame(data = pca.fit_transform(X_train_prep)
                          , columns = ['principal component '+ str(i) for i in range(1,2+1)])

for i, c, s in (zip(range(len(set(y_train_prep.to_numpy().reshape(-1,)))), ['b', 'g'], ['o', '^'])):
    ix = y_train_prep.to_numpy().reshape(-1,) == i
    plt.scatter(X_train1.iloc[:, 0][ix], X_train1.iloc[:, 1][ix], color=c, marker=s, s=10, label=list(set(y_train_prep.to_numpy().reshape(-1,)))[i])
plt.legend(loc=2, scatterpoints=1)
plt.xlabel("feature 1-" + X_train1.columns[0])
plt.ylabel("feature 2-" + X_train1.columns[1])
plt.show()
```

```python
pca = PCA(n_components=3)
X_train2 = pd.DataFrame(data = pca.fit_transform(X_train_prep)
                          , columns = ['principal component '+ str(i) for i in range(1,3+1)])

fig = plt.figure(figsize=(15,20))
ax = fig.add_subplot(projection='3d')
for i, c, s in (zip(range(len(set(y_train_prep.to_numpy().reshape(-1,)))), ['b', 'g'], ['o', '^'])):
    ix = y_train_prep.to_numpy().reshape(-1,) == i
    ax.scatter(X_train2.iloc[:, 0][ix], X_train2.iloc[:, 1][ix], X_train2.iloc[:, 2][ix], color=c, marker=s,s =20, label=list(set(y_train_prep.to_numpy().reshape(-1,)))[i])
ax.legend(loc=2, scatterpoints=1)
ax.set_xlabel("feature 1-" + X_train2.columns[0])
ax.set_ylabel("feature 2-" + X_train2.columns[1])
ax.set_zlabel("feature 3-" + X_train2.columns[2])

plt.show()
```

# Data tuning

## Tuning by HalvingGridSearchCV

```python
f2_scorer = make_scorer(fbeta_score,beta = 2)
model = SVC()
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [10,1,0.1,0.01],'kernel': ['rbf', 'poly', 'linear']}
grid = HalvingGridSearchCV(model,param_grid,refit=True,verbose=False, scoring = f2_scorer,cv = 5)
grid.fit(X_train_prep,y_train_prep)
print(grid.best_estimator_)
print(grid.best_params_)
grid_predictions = grid.predict(X_val)
print("f2_score: "+ str(fbeta_score(y_val,grid_predictions, beta =2 )))
print(confusion_matrix(y_val,grid_predictions))
print(classification_report(y_val,grid_predictions))
```

## Hand-Tuning 

```python
model = SVC(C =0.1, gamma = 0.01, kernel = 'rbf')
model.fit(X_train_prep, y_train_prep)
Y_predict = model.predict(X_val)
print(accuracy_score(y_val, Y_predict))
print(fbeta_score(y_val,Y_predict, beta = 2))
print(fbeta_score(y_train_prep,model.predict(X_train_prep), beta = 2))
print(confusion_matrix(y_val,Y_predict))
print(classification_report(y_val,Y_predict))
```

# Model

```python
model = SVC(C =0.1, gamma = 0.01, kernel = 'rbf')
model.fit(X_train_prep, y_train_prep)
Y_predict = model.predict(X_test)
print(accuracy_score(y_test, Y_predict))
print(fbeta_score(y_test,Y_predict, beta = 2))
print(fbeta_score(y_train_prep,model.predict(X_train_prep), beta = 2))
print(confusion_matrix(y_test,Y_predict))
print(classification_report(y_test,Y_predict))
```

# ROC_CURVE

```python
plot_roc_curve(model, X_test, y_test)
```

# Learning curve for misclassification rate over data size

```python
from sklearn.model_selection import learning_curve

X = pd.concat([X_train_prep, X_test_prep])
y = pd.concat([y_train_prep,y_test_prep])
train_sizes, train_scores, test_scores= learning_curve(model, X , y,train_sizes = np.linspace(0.1, 1.0, 10), cv = 5, random_state =1, scoring = 'accuracy')
plt.plot(train_sizes,1-np.mean(train_scores,axis=1), label = "Training")
plt.plot(train_sizes,1-np.mean(test_scores,axis=1), label = 'Validation')
plt.legend(loc = "upper right")
```
